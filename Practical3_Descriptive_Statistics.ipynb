Measures of Central Tendency and Variability

## Import Libraries

```
import pandas as pd
import numpy as np
```

## Create Sample Dataset

```
data = {
    'Age_Group': ['Young','Young','Adult','Adult','Senior','Senior','Young','Adult','Senior','Young'],
    'Income': [20000, 22000, 50000, 52000, 30000, 32000, 21000, 51000, 31000, 23000]
}
df = pd.DataFrame(data)
df
```

## Summary Statistics Grouped by Age Group

```
group_stats = df.groupby('Age_Group')['Income'].agg(['mean','median','min','max','std'])
group_stats
```

## Load Iris Dataset

```
from sklearn.datasets import load_iris
iris = load_iris(as_frame=True)
iris_df = iris.frame
iris_df
```

## Statistical Details for Each Species

```
iris_df.groupby('target').agg(['mean','std','min','max'])
```

## Percentile Calculation

```
np.percentile(iris_df['sepal length (cm)'], [25,50,75])
```
