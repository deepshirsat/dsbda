```python
# Practical 8: Data Visualization using Titanic Dataset

This notebook:
1. Explores Titanic dataset using Seaborn
2. Identifies patterns in passenger data
3. Visualizes fare distribution using histogram
```

```python
## Step 1: Import Libraries
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Step 2: Load Inbuilt Titanic Dataset
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
## Step 3: Dataset Overview
```

```python
df.info()
df.describe()
```

```python
## Step 4: Visualizing Patterns in Data
```

```python
# Survival count
sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.show()

# Survival based on gender
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival by Gender")
plt.show()

# Survival based on passenger class
sns.countplot(x='pclass', hue='survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```
